Linear Regression


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
data = pd.read_csv('/content/hw_200.csv')  # Add dataset path here

# Clean up column names by stripping spaces and removing extra characters
data.columns = data.columns.str.strip().str.replace('"', '')

# Print the cleaned column names to verify
print(data.columns)

# Replace 'Height' and 'Weight' with the actual cleaned column names
X = data[['Height(Inches)']].values  # Feature (independent variable)
y = data[['Weight(Pounds)']].values  # Target (dependent variable)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot the original data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(X, y_pred, color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('Height (Inches)')
plt.ylabel('Weight (Pounds)')
plt.title('Simple Linear Regression')
plt.legend()

# Show the plot
plt.show()

```

Calculate the R-squared value of the model and interpret it.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load the dataset (replace 'your_dataset.csv' with your actual file path)
data = pd.read_csv('/content/hw_200.csv')  # Add dataset path here

# Clean up column names by stripping spaces and removing extra characters
data.columns = data.columns.str.strip().str.replace('"', '')

# Replace 'Height(Inches)' and 'Weight(Pounds)' with the actual cleaned column names
X = data[['Height(Inches)']].values  # Feature (independent variable)
y = data[['Weight(Pounds)']].values  # Target (dependent variable)

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Calculate the R-squared value
r_squared = r2_score(y, y_pred)

# Print the R-squared value
print(f"R-squared: {r_squared}")

# Plot the original data points
plt.scatter(X, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(X, y_pred, color='red', label='Regression Line')

# Adding labels and title
plt.xlabel('Height (Inches)')
plt.ylabel('Weight (Pounds)')
plt.title('Simple Linear Regression')
plt.legend()

# Show the plot
plt.show()

```

Plot the residuals and check for any patterns.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Load your dataset
data = pd.read_csv("/content/hw_200.csv")  # Replace with the correct dataset path

# Clean the column names by stripping spaces or unwanted characters
data.columns = data.columns.str.strip().str.replace('"', '').str.replace(' ', '_')

# Access the cleaned columns
X_simple = data[['Height(Inches)']].values  # Independent variable (Height)
y = data[['Weight(Pounds)']].values  # Dependent variable (Weight)

# Perform Linear Regression without feature scaling
model = LinearRegression()
model.fit(X_simple, y)

# Make predictions
y_pred = model.predict(X_simple)

# Calculate residuals
residuals = y - y_pred

# Plotting the residuals against the predicted values
plt.scatter(y_pred, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residual Line')
plt.xlabel('Predicted Weight')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.legend()
plt.show()

# Alternatively, you can plot residuals against the independent variable (Height)
plt.scatter(X_simple, residuals, color='green', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residual Line')
plt.xlabel('Height(Inches)')
plt.ylabel('Residuals')
plt.title('Residuals vs Height')
plt.legend()
plt.show()

```

Compare the performance of linear regression with and without feature
scaling.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Load your dataset
data = pd.read_csv("/content/hw_200.csv")  # Replace with the correct dataset path

# Clean the column names by stripping spaces or unwanted characters
data.columns = data.columns.str.strip().str.replace('"', '').str.replace(' ', '_')

# Check the cleaned column names
print(data.columns)

# Now you can access the columns without errors
X_simple = data[['Height(Inches)']].values  # Independent variable (Height)
y = data[['Weight(Pounds)']].values  # Dependent variable (Weight)

# Perform Linear Regression without feature scaling
model = LinearRegression()
model.fit(X_simple, y)

# Make predictions
y_pred = model.predict(X_simple)

# Plotting the regression line
plt.scatter(X_simple, y, color='blue', label='Data Points')
plt.plot(X_simple, y_pred, color='red', label='Regression Line')
plt.xlabel('Height(Inches)')
plt.ylabel('Weight(Pounds)')
plt.title('Height vs Weight (Linear Regression)')
plt.legend()
plt.show()

# Print the R-squared value
print(f"R-squared: {model.score(X_simple, y):.2f}")

# Perform Linear Regression with feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_simple)

# Fit the scaled model
model_scaled = LinearRegression()
model_scaled.fit(X_scaled, y)

# Make predictions for the scaled model
y_pred_scaled = model_scaled.predict(X_scaled)

# Plotting the regression line for scaled features
plt.scatter(X_simple, y, color='blue', label='Data Points')
plt.plot(X_simple, y_pred_scaled, color='green', label='Scaled Regression Line')
plt.xlabel('Height(Inches)')
plt.ylabel('Weight(Pounds)')
plt.title('Height vs Weight (Scaled Linear Regression)')
plt.legend()
plt.show()

# Print the R-squared value for the scaled model
print(f"R-squared (scaled): {model_scaled.score(X_scaled, y):.2f}")

```
